```python
#import modules
#code taken from week 1 notes

import pandas as pd #this package is used for dataframes
from bs4 import BeautifulSoup
import requests
pd.set_option('display.max_colwidth', 150) #important for getting all the text
```

```python
#first source: scrape a forum for the F-150 truck
#code taken from week 1 notes

#the page used is the 2015-2020 F-150 truck forum
page = requests.get('https://www.f150forum.com/f118/')
soup = BeautifulSoup(page.text, "lxml")
#print(soup)
```

```python
#the following pieces of code will grab the information rough information we want, it will be cleaned later
#most of the code taken from week 1 notes

#import the re package - this allows for regular expressions
import re

#grab all of the thread titles on the main page
#some code taken from https://stackoverflow.com/questions/14257717/python-beautifulsoup-wildcard-attribute-id-search
BSobj = soup.find_all(id=re.compile("thread_title"))

#test print lines

#print(BSobj)
#print()
#print("We scraped", len(BSobj), "threads")
#print()

#grab the last update timestamp of each thread
BSobj2 = soup.find_all("div",{"class":"smallfont text-right"})

#test print lines

#print(BSobj2)
#print()
#print("We scraped", len(BSobj2), "threads")
#print()

#grab the reply and view count for each thread
BSobj3 = soup.find_all("div",{"class":"tcell alt2 smallfont"})

#test print lines

#print(BSobj3)
#print()
#print("We scraped", len(BSobj3), "threads")
```

```python
#import time package, this lets us work with times/dates
import time

#pre-define lists and dictionary to put cleaned data
stamps = []
correct_stamps = []
views = []
correct_views = []
threads = {}

#put soup objects into lists
#some code taken from https://stackoverflow.com/questions/44502482/how-to-create-and-fill-a-list-of-lists-in-a-for-loop

#first, we will simply fill the lists with the soup objects
#some code taken from https://stackoverflow.com/questions/12124191/beautifulsoup-text-method-returns-text-without-separators-n-r-etc
for info in BSobj2:
    stamps.append(info.get_text(separator=" ",strip=True))

#some code taken from https://www.crummy.com/software/BeautifulSoup/bs4/doc/#attributes
for count in BSobj3:
    views.append(count.attrs['title'])

#next, iterate through the lists while cleaning the data
#some code taken from https://medium.com/@Alexander_H/removing-characters-before-after-and-in-the-middle-of-strings-fb4930cce76a
#this will split the soup-formatted information and keep only the wanted piece
x=0
for what in stamps:
    splitter = stamps[x].split(" by ")
    correct_stamps.append(splitter[0])
    x=x+1

x=0
for what in views:
    splitter = views[x].split(": ",2)
    correct_views.append(splitter[2])
    x=x+1

#finally, combine the thread, timestamp, and count info into one dictionary
#also add timestamp for today as well as the forum type
x=0
for title in BSobj:
    threads.setdefault(title.get_text(strip=True),[])
    threads[title.get_text(strip=True)].append(correct_stamps[x])
    threads[title.get_text(strip=True)].append(correct_views[x])
    threads[title.get_text(strip=True)].append(time.strftime("%m/%d/%Y"))
    threads[title.get_text(strip=True)].append("2015-2020 F-150")
    x=x+1
```

```python
#this is a test print to make sure the dictionary looks as expected
#print(threads)
```

```python
#move the dictionary data into a pandas dataframe
#code taken from week 1 notes

df1 = pd.DataFrame.from_dict(threads,orient="index")
df1.reset_index(level=[0], inplace=True)
#print(df1.shape)

df1.columns = ['Thread Title', 'Last Update', 'View Count', 'Date Accessed', 'Category']
df1.head()
```

```python
#second source: scrape a parts shop for F-150 trucks
#code taken from week 1 notes

#the page used is the 2015-2020 F-150 Leveling Kits
page2 = requests.get(url='https://www.stage3motorsports.com/2015-2020-f150-spacer-leveling-kits.html')
soup2 = BeautifulSoup(page2.text, "lxml")
#print(soup2)
```

```python
#code taken from week 1 notes

#grab all products listed on the page
BSobj4 = soup2.find_all("div",{"class":"name"})
#print(BSobj4)
#print()
#print("We scraped", len(BSobj4), "categories")
#print()

#grab prices for all products on the page
BSobj5 = soup2.find_all("span",{"class":"regular-price"})
#print(BSobj5)
#print()
#print("We scraped", len(BSobj5), "categories")
#print()

#grab average review for all products on the page
BSobj6 = soup2.find_all("img",{"alt":"Average Rating"})
#print(BSobj6)
#print()
#print("We scraped", len(BSobj6), "categories")
```

```python
#pre-define lists and dictionary to put cleaned data
prices = []
correct_prices = []
reviews = []
correct_reviews = []
kits = {}

#put soup objects into lists
#some code taken from https://stackoverflow.com/questions/44502482/how-to-create-and-fill-a-list-of-lists-in-a-for-loop

#first, we will simply fill the lists with the soup objects
for cost in BSobj5:
    prices.append(cost.get_text(strip=True))
    
#print(prices)
#print()

#use "slices" to remove all even elements of the list
#some code taken from https://stackoverflow.com/questions/28883769/remove-odd-indexed-elements-from-list-in-python
correct_prices = prices[1::2]

#print(correct_prices)

#put raw rating data into a list
for rating in BSobj6:
    reviews.append(rating.attrs["data-src"])
    
#print(reviews)

#clean rating data to get only the rating value
#some code taken from https://thispointer.com/python-how-to-get-last-n-characters-in-a-string/
x=0
for what in reviews:
    splitter = reviews[x].split(".")
    #print(splitter[0][-1]) 
    correct_reviews.append(splitter[0][-1])
    x=x+1
    
#print(correct_reviews)

#put the product info plus all of the cleaned data into a dictionary
x=0
for product in BSobj4:
    kits.setdefault(product.get_text(strip=True),[])
    kits[product.get_text(strip=True)].append(correct_prices[x])
    kits[product.get_text(strip=True)].append(correct_reviews[x])
    kits[product.get_text(strip=True)].append(time.strftime("%m/%d/%Y"))
    kits[product.get_text(strip=True)].append("F-150 Spacer Leveling Kits")
    x=x+1
    
#print(len(kits))
```

```python
#move the dictionary data into a pandas dataframe
#code taken from week 1 notes

df2 = pd.DataFrame.from_dict(kits,orient="index")
df2.reset_index(level=[0], inplace=True)
#print(df2.shape)

df2.columns = ['Product', 'Price', 'Average Review (out of 5)', 'Date Accessed', 'Category']
df2.head()
```
